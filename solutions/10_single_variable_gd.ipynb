## Phase: Active Learning (ipywidgets + matplotlib)
Let's make the learning rate interactive! Drag the slider to see how changing the learning rate changes the gradient descent path.

```
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, FloatSlider

def plot_gd(alpha=0.1):
    x = np.linspace(-5, 5, 100)
    y = x**2  # Example function
    plt.plot(x, y, label='f(x) = x^2')
    
    # Scaffold for plotting steps (x_steps, y_steps)
    # plt.scatter(x_steps, y_steps, color='red')
    # plt.plot(x_steps, y_steps, color='red', linestyle='--')
    
    plt.title(f'Gradient Descent Path (alpha={alpha})')
    plt.legend()
    plt.show()

interact(plot_gd, alpha=FloatSlider(value=0.1, min=0.01, max=1.0, step=0.05));
```
