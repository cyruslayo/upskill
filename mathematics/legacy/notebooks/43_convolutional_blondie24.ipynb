---## Phase −1 — Theoretical Foundation### The Spatial Problem> 📖 *From Algorithms Ch. 43:* In 2001, Fogel & Chellapilla improved Blondie24 by adding a **Convolutional Layer**. This acts as a spatial preprocessing layer, allowing the network to perceive spatial characteristics of the board at different levels of zoom.In a standard MLP, the $8 \times 8$ checkers board is flattened into a 64-element 1D list. The network completely loses the 2D geometric structure! It doesn't know that square 1 is "above" square 9.### The Convolutional ConceptA convolutional layer fixes this by processing 2D patches of the board.Blondie24 defined subnetworks that look at every possible $3 \times 3$, $4 \times 4$, $5 \times 5$, $6 \times 6$, $7 \times 7$, and $8 \times 8$ subsquare of the board!- For an $8 \times 8$ board, there are 36 possible $3 \times 3$ subsquares.- Each of these 36 patches has its own node in the Convolutional Layer.This structure mathematically forces the neural network to "see" localized 2D patterns (like an opponent setting up a jump) before making decisions!

### Comprehension Check ✅Why did the addition of the convolutional layer increase the number of weights in Blondie24 from 1,742 to 5,047?<details><summary>💡 Explanation after a retrieval attempt</summary>Because the input layer now connects to 92 different convolutional nodes ($1 + 4 + 9 + 16 + 25 + 36 + 1$), and each connection processes a 2D patch of inputs, requiring vastly more weight matrices to filter the visual information.</details>
