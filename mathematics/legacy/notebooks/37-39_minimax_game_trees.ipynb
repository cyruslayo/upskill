---## Phase −1 — Theoretical Foundation### Game Trees> 📖 *From Algorithms Ch. 37:* A game tree represents all possible moves in a game. The current board is the "root". Each possible legal move creates a "child" branch, leading to a new board state. We can explore the tree to find winning paths.### The Minimax AlgorithmIf we explore the tree all the way down to a Win (+1), Loss (-1), or Tie (0), we can work backwards to figure out the best move right now. - When it is **Our Turn**, we want to maximize our score. We will choose the child branch with the MAXIMUM score.- When it is the **Opponent's Turn**, we must assume they play perfectly. They want to minimize our score. They will choose the child branch with the MINIMUM score.By recursively calling Minimax down the tree, we alternatingly take the Max and Min of the children to find the true, guaranteed value of any game state!
