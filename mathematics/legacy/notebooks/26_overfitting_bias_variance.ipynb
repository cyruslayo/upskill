---## Phase −1 — Theoretical Foundation### The Problem with Training Accuracy> 📖 *From Algorithms Ch. 26:* If we define “accuracy” as simply matching with points in the dataset, then an 8th-degree polynomial is incredibly accurate. But it has to curve and contort itself in unrealistic ways between data points.Imagine tracking a rocket's height over time. The sensor data is noisy.- A **Linear Regression** draws a perfectly straight line. It completely ignores that the rocket is accelerating upwards. This model is too rigid. It **Underfits** the data. This is an error due to **Bias** (the model stubbornly assumes the world is simple).- An **8th-Degree Polynomial** hits every single sensor point perfectly! But between seconds 9 and 10, the polynomial predicts the rocket shoots up to space, then crashes into the ground, then shoots back up. It trusts the noisy sensor data too much. This model is too flexible. It **Overfits** the data. This is an error due to **Variance** (the model drastically changes its shape depending on the exact points in the dataset).### The Solution: Cross ValidationThe true test of a model is how well it predicts data it has *never seen before*. Instead of checking accuracy on the training data, we use **Leave-One-Out Cross Validation**:1. Remove 1 point from the dataset.2. Train the model on the remaining $N-1$ points.3. Ask the model to predict the 1 point we removed. Record the error.4. Put the point back, and repeat for every point in the dataset!The sum of these errors is the true measure of generalization.

### Comprehension Check ✅Anthropomorphizing models can help intuition. The textbook says a high-bias model is "dumb" (cannot comprehend complexity). What is the analogy for a high-variance model?<details><summary>💡 Explanation after a retrieval attempt</summary>A high-variance model is "paranoid": it thinks it is seeing all sorts of complicated relationships that just aren't there (it treats random noise as ground truth).</details>
