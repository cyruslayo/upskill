---## Phase −1 — Theoretical Foundation### Matrices as Transformations> 📖 *From Linear Algebra §3.1:* Matrices can be thought of as objects that transform vectors.A matrix is a grid of numbers. An $m \times n$ matrix has $m$ rows and $n$ columns.If we multiply an $m \times n$ matrix by an $n$-dimensional vector, we get an $m$-dimensional vector.$$\begin{bmatrix} a & b \\ c & d \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = x \begin{bmatrix} a \\ c \end{bmatrix} + y \begin{bmatrix} b \\ d \end{bmatrix} = \begin{bmatrix} ax + by \\ cx + dy \end{bmatrix}$$Notice that the columns of the matrix dictate where the basis vectors $\langle 1, 0 \rangle$ and $\langle 0, 1 \rangle$ land!### Matrix Multiplication> 📖 *From Linear Algebra §3.2:* When multiplying two matrices $A$ and $B$, the result $AB$ represents applying transformation $B$ first, then transformation $A$.The element in row $i$ and column $j$ of the product matrix $C = AB$ is the **dot product** of the $i$-th row of $A$ and the $j$-th column of $B$.$$C_{ij} = \sum_{k=1}^n A_{ik} B_{kj}$$**Important:** Matrix multiplication is NOT commutative. $AB \neq BA$ in general. Order matters!

```python
# Symbolic matrix multiplicationimport sympy as spa, b, c, d, x, y, z, w = sp.symbols('a b c d x y z w')A = sp.Matrix([[a, b], [c, d]])B = sp.Matrix([[x, y], [z, w]])print('AB =')sp.pprint(A * B)print('\nBA =')sp.pprint(B * A)print('\nNotice that AB != BA')
```
