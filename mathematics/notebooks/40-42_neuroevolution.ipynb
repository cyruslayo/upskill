---## Phase −1 — Theoretical Foundation### The Problem with Gradient Descent in Games> 📖 *From Algorithms Ch. 42:* When training an AI to play Checkers, we don't have a dataset of "perfect moves". And we can't easily backpropagate through thousands of discrete game tree decisions.### NeuroevolutionInstead of calculating gradients mathematically, we use **Darwinian Evolution**!1. **Initialize** 15 random Neural Networks.2. **Mutate**: Each network generates a "child" with slightly randomized weights.3. **Evaluate**: We now have 30 networks. We make them play Checkers against each other in a massive tournament! We track how many games each network wins.4. **Select**: We kill the 15 lowest-scoring networks. The 15 best networks survive.5. Go back to step 2 and repeat for 800 generations!This exact technique was used by Fogel & Chellapilla in 1999 to create *Blondie24*, a Checkers AI that taught itself to achieve an expert rating on an online gaming site, entirely through self-play!

### Comprehension Check ✅Why do we mutate the weights randomly instead of trying to calculate which direction is best?<details><summary>💡 Answer</summary>Because the "fitness landscape" of a complex game is highly jagged, and we have no mathematical formula to compute the gradient of "winning a game of checkers". Random mutation, combined with strict selection, effectively acts as a "stochastic gradient descent" that explores the landscape blindly but relentlessly moves uphill toward mastery.</details>
