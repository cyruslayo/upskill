---## Phase −1 — Theoretical Foundation### The Intuition> 📖 *From Algorithms Ch. 10:* Gradient descent is a technique for minimizing differentiable functions. We take an initial guess, and repeatedly use the gradient to nudge that guess further and further “downhill” into an actual minimum.If the derivative $f'(x)$ is positive, the function is sloping *up*. To go downhill, we must move left (decrease $x$).If the derivative is negative, the function is sloping *down*. To go downhill, we must move right (increase $x$).### The Update Rule$$x_{n+1} = x_n - \alpha f'(x_n)$$Where:- $x_n$ is the current guess- $f'(x_n)$ tells us the direction and steepness of the slope- $\alpha$ is the **learning rate** (usually $\approx 0.01$), controlling how large of a step we take.We subtract $\alpha f'(x_n)$ because we always want to move in the *opposite* direction of the slope.

### Comprehension Check ✅1. What happens if $\alpha$ is too large (e.g. 1.5)?2. What happens if $\alpha$ is too small (e.g. 0.000001)?<details><summary>💡 Answers</summary>1. The guess will "overshoot" the minimum, jumping back and forth across the valley, potentially diverging to infinity.2. The guess will update by a miniscule amount each time, requiring an excessive number of iterations to converge.</details>
