---## Phase −1 — Theoretical Foundation### The Need for Speed> 📖 *From Algorithms Ch. 13:* With an average-case time complexity of $O(n \log n)$, Merge Sort and Quicksort are generally faster than selection, bubble, and insertion sort. And unlike counting sort, they don't blow up in memory on large numbers.### Merge Sort (Divide & Conquer)1. Break the array in half.2. Recursively call Merge Sort on each half.3. Merge the two sorted halves back together.Base Case: Arrays of length 0 or 1 are already sorted.### Quicksort1. Pick a random `pivot` element.2. Split the rest of the array into three arrays: elements `< pivot`, `== pivot`, and `> pivot`.3. Recursively call Quicksort on the `<` array and `>` array.4. Concatenate the sorted `<` array, the `==` array, and the sorted `>` array.

### Chalkboard DefenseExplain conceptually why Merge Sort takes $O(N \log N)$ time. <details><summary>💡 Sketch</summary>The recursive splitting forms a tree. An array of size $N$ can be split in half $\log_2 N$ times before hitting size 1. So the tree has $\log_2 N$ levels.At each level, we merge sub-arrays. Merging two arrays of size $K$ takes $O(K)$ time. If you sum up the merge times across an entire level of the tree, you are merging $N$ total elements, which takes $O(N)$ time.Since there are $\log N$ levels, and each level does $O(N)$ work, the total time is $O(N \log N)$.</details>---📚 **Next:** Algorithms 14 (Basic Matrix Arithmetic)
