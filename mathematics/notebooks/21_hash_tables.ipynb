---## Phase −1 — Theoretical Foundation### Dictionaries are Hash Tables> 📖 *From Algorithms Ch. 21:* Under the hood, dictionaries are hash tables. The most elementary (and inefficient) version of a dictionary is just a list of `(key, value)` tuples.If we have 1 million tuples, looking up a key requires $O(N)$ time. We must scan the whole list.### The Solution: Buckets and Hash FunctionsInstead of 1 giant list, what if we have 100 separate lists ("buckets")? We use a **Hash Function** that takes the string key, does some math on it, and spits out an integer from $0$ to $99$. Now, when we insert `("apple", 5)`, the hash function says `hash("apple") = 42`. We instantly go to bucket 42 and append it there.When we want to look up `"apple"`, we hash it, get 42, go to bucket 42, and search that tiny list. Time complexity drops to $O(1)$!

### Comprehension Check ✅1. What is a "Collision" in a hash table?2. What happens if our Hash Function is terribly designed and always outputs `0` regardless of the string?<details><summary>💡 Answers</summary>1. A collision happens when two completely different keys (e.g. "apple" and "banana") happen to yield the exact same hash integer, and thus end up in the exact same bucket.2. Every single key gets put into Bucket 0. The other 99 buckets remain empty. You essentially regress back to a single giant list, and performance drops to $O(N)$. This is why uniform hashing is critical!</details>

What "attack" does this prevent?<details><summary>💡 Sketch</summary>If the hash algorithm is deterministic and public, a malicious user could intentionally generate 100,000 different string keys that *all* collide to Bucket 0. If they send a JSON payload with these keys to a web server, the server's Hash Table degrades to $O(N^2)$ insertion time, causing the server's CPU to max out and crash.
