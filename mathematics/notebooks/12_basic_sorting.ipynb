---## Phase −1 — Theoretical Foundation### $O(n^2)$ Sorting Algorithms> 📖 *From Algorithms Ch. 12:* There are many methods for sorting arrays. The simplest are Selection, Bubble, and Insertion sort. They all take $O(n^2)$ time.- **Selection Sort:** Find the minimum element in the remaining array, and move it to the end of the sorted array.- **Bubble Sort:** Repeatedly swap adjacent elements if they are out of order. In each pass, the largest remaining element "bubbles" to the top. Stop when a full pass makes 0 swaps.- **Insertion Sort:** Take the next out-of-order element and swap it backwards until it sits in its correct sorted position among the already-processed elements.### Counting Sort: An $O(N+K)$ AlternativeCounting sort does not compare elements! It counts occurrences of each number, then generates a new array directly from the counts. It is extremely fast but can use a massive amount of memory if the range of numbers is very large.

### Comprehension Check ✅1. Why is Bubble Sort $O(n^2)$?2. If an array contains the numbers `[1, 1000000]`, why is Counting Sort a terrible idea?<details><summary>💡 Answers</summary>1. In the worst case (a reverse-sorted array), it takes $n$ passes, and each pass performs $n$ comparisons/swaps. $n \times n = n^2$.2. Counting sort requires a `counts` array of size `max - min + 1`. You would allocate an array of 1,000,000 integers just to sort 2 numbers!</details>
