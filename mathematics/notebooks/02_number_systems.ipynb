---## Phase −1 — Theoretical Foundation### Number Systems> 📖 *From Algorithms Ch. 2:* We are used to representing numbers using ten characters (0-9). This is called the **decimal** (base-10) system. The **binary** (base-2) system uses two characters (0, 1), and the **hexadecimal** (base-16) system uses sixteen characters (0-9, A-F, where A=10, ..., F=15).### Converting from Base-b to DecimalIf you have a number $x_n x_{n-1} \dots x_1 x_0$ in base-$b$, its decimal value is:$$x_n \cdot b^n + x_{n-1} \cdot b^{n-1} + \dots + x_1 \cdot b^1 + x_0 \cdot b^0$$**Example:** `11010` in binary$= 1(2^4) + 1(2^3) + 0(2^2) + 1(2^1) + 0(2^0) = 16 + 8 + 0 + 2 + 0 = 26$.### Converting from Decimal to Base-bRepeatedly compute the highest power of $b$ that is $\leq$ your number, subtract off the largest multiple of that power, and record the multiple as your digit.*(Equivalently, you can repeatedly divide by $b$ and record the remainders, then reverse them).*
