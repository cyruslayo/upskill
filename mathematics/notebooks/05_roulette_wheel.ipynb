---## Phase −1 — Theoretical Foundation### Non-Uniform Distributions> 📖 *From Algorithms Ch. 5:* Flipping a fair coin is sampling from `[0.5, 0.5]`. What if we want to roll a die with 2 red, 1 blue, 1 green, and 1 yellow face? We must sample from `[0.4, 0.2, 0.2, 0.2]`.### Roulette Wheel SelectionTo sample an index from an arbitrary distribution:1. Turn the distribution into a **cumulative distribution**. (e.g. `[0.4, 0.6, 0.8, 1.0]`)2. Sample a random float $r$ between 0 and 1. (e.g. `r = 0.63`)3. Find the **first index** in the cumulative distribution that is $\geq r$. (0.8 is the first value $\geq$ 0.63, so we pick index 2).
