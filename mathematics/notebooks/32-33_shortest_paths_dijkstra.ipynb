---## Phase −1 — Theoretical Foundation### Weighted Graphs> 📖 *From Algorithms Ch. 32:* In the real world, roads have speed limits and distances. Not all edges are equal! A weighted graph assigns a numerical cost to each edge.In BFS, we assumed every edge had a cost of 1. To find the shortest path in a weighted graph, we must use **Dijkstra's Algorithm**.### Dijkstra's AlgorithmDijkstra's algorithm is essentially BFS, but instead of a standard Queue, it uses a **Priority Queue**.1. We maintain a table of the shortest known `distance` to every node (initialized to infinity).2. We push `(0, start_node)` into the priority queue.3. While the queue is not empty, we pop the node with the *smallest distance*.4. For each neighbor, we calculate `new_dist = current_dist + edge_weight`.5. If `new_dist` is less than the neighbor's known `distance`, we update the table, set `previous[neighbor] = current_node`, and push `(new_dist, neighbor)` into the priority queue.Because we always process the *closest* unvisited node, once we pop a node, we are mathematically guaranteed that we have found its absolute shortest path!

### Comprehension Check ✅Why can't Dijkstra's algorithm handle graphs with *negative* edge weights?<details><summary>💡 Explanation after a retrieval attempt</summary>Dijkstra assumes that once a node is popped from the priority queue, its shortest path is permanently finalized. If there are negative weights, a longer path with more edges could suddenly drop in total cost, invalidating the "finalized" distances. Algorithms like Bellman-Ford are required for negative weights.</details>
