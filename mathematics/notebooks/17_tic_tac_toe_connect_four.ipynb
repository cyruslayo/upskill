### Chalkboard DefenseExplain how you would write a `MinimaxPlayer` that never loses at Tic-Tac-Toe. What does the algorithm do conceptually?<details><summary>💡 Sketch</summary>Minimax generates a **Game Tree** of every possible future move. At the bottom of the tree, it scores wins as +1, losses as -1, and ties as 0.It then percolates these scores back up the tree. When it is X's turn, X chooses the move that *maximizes* the score. When it is O's turn, O chooses the move that *minimizes* the score.Because the search space of Tic-Tac-Toe is tiny ($< 362,880$ nodes), Minimax can see the end of the game from turn 1 and play perfectly.</details>---📚 **Next:** Algorithms 18 (Euler Estimation)
