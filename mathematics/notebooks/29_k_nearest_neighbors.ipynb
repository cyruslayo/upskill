---## Phase −1 — Theoretical Foundation### The K-Nearest Neighbors Algorithm> 📖 *From Algorithms Ch. 29:* Given a data set of points labeled with classes, KNN predicts the class of a new point by finding the $K$ points in the dataset nearest to the new point, and predicting the majority class of those neighbors.Imagine we have a dataset of cookies: `(portion_butter, portion_sugar, type)`.Type 0 is Shortbread, Type 1 is Sugar Cookie.If we have a new cookie with `(0.25, 0.30)`, we compute the Euclidean distance $\sqrt{(x_2-x_1)^2 + (y_2-y_1)^2}$ to every cookie in our dataset. If $K=3$, we look at the 3 closest cookies. If 2 are Shortbread and 1 is Sugar, we predict Shortbread!### Choosing KWe use **Leave-One-Out Cross Validation** (from Chapter 26). For every point in the dataset, we pretend we don't know its label, and try to predict it using the *rest* of the dataset. We sum up the accuracy for each $K$, and pick the $K$ with the highest accuracy!

### Comprehension Check ✅Why is it mathematically dangerous to use KNN on a dataset where `Butter` is measured in tiny decimals like $0.05$ cups, but `Sugar` is measured in huge integers like $300$ grams?<details><summary>💡 Answer</summary>Because the Euclidean distance formula treats all axes equally! A difference of $10$ grams of sugar will absolutely overwhelm a difference of $0.1$ cups of butter in the distance formula. The algorithm will essentially ignore the butter. We must **Normalize** all variables to a $0-1$ scale before running KNN.</details>
