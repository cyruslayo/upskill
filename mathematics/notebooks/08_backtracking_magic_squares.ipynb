---## Phase −1 — Theoretical Foundation### The Problem with Brute Force> 📖 *From Algorithms Ch. 8:* Brute force search can often be slow or infeasible when there are lots of possibilities. However, **backtracking** can drastically cut down the number of possibilities.We want to find a $3 \times 3$ **magic square** using the digits 1-9 without repetition. A magic square has all rows, columns, and diagonals summing to exactly 15.If we use 9 nested loops, we check $9! = 362,880$ arrangements.But if we fill the top row with `[1, 2, 3]`, the sum is 6. This row will *never* sum to 15, no matter what we put in the other 6 squares! ### BacktrackingWhenever we reach an arrangement that can no longer become valid, we **skip** exploring it further. We use a `continue` statement to jump to the next iteration of the loop, effectively pruning thousands of hopeless branches at once.
