---## Phase −1 — Theoretical Foundation### The Problem with Pseudoinverse Transforms> 📖 *From Algorithms Ch. 27:* If we transform $y = a x^b$ using logarithms to apply the pseudoinverse, we find the optimal fit for the *transformed* equation, which is often NOT the optimal fit for the original equation! We must use Gradient Descent instead.### Gradient Descent for RegressionTo fit a model using gradient descent, we construct an expression representing the error (RSS): $$RSS = \sum (pred - y_{actual})^2$$For a power regression $y = a \cdot x^b$, the predicted value is $a \cdot x^b$.We take the partial derivative of RSS with respect to $a$ and $b$, and take small steps $-\alpha \nabla RSS$ to minimize the error!### Multiple Variables & Interaction TermsSuppose $x_1$ is "scoops of peanut butter" and $x_2$ is "slices of roast beef". A logistic model $y = \frac{1}{1 + e^{-(a_1x_1 + a_2x_2 + b)}}$ predicts if someone likes the sandwich.If both ingredients are good individually, $a_1 > 0, a_2 > 0$. But if you combine PB & Beef, it tastes awful! Linear models cannot capture this. We add an **Interaction Term** $x_1 x_2$:$$z = a_1x_1 + a_2x_2 + a_{12}(x_1x_2) + b$$The algorithm will learn a *negative* weight for $a_{12}$ to drop the probability if both are present!

### Comprehension Check ✅Why do interaction terms vanish unless both variables are non-zero?<details><summary>💡 Answer</summary>Because they are multiplied together ($x_1 \cdot x_2$). If there is no peanut butter ($x_1 = 0$), then $0 \cdot x_2 = 0$, so the interaction term completely disappears and has no effect on the prediction. It only activates when BOTH ingredients are present.</details>
