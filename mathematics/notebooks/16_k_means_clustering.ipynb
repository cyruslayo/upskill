---## Phase −1 — Theoretical Foundation### The K-Means Algorithm> 📖 *From Algorithms Ch. 16:* K-means clustering groups similar records together. It guesses initial clusters, and repeatedly updates them to make them more cohesive.1. **Initialize:** Randomly divide the data into $K$ parts (clusters). Compute the center (mean) of each part.2. **Assign:** Loop through all data points. Re-assign each point to the cluster whose center is closest (using Euclidean distance).3. **Update:** Compute the new center for each cluster by taking the mean of all points newly assigned to it.4. **Repeat:** Repeat Assign & Update until the cluster assignments stop changing.### The Elbow MethodHow do we choose $K$? If $K = N$ (number of points), every point is its own cluster (distance=0, but useless). We plot the **Total Distance** (sum of distances from each point to its cluster center) against $K$. The curve goes down. We pick the "elbow" of the curve — the point of diminishing returns.
