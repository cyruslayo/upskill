---## Phase −1 — Theoretical Foundation### What is a Recursive Sequence?> 📖 *From Algorithms Ch. 3:* A **recursive sequence** is a sequence where each term is a function of the previous terms.Example: Starting with 3, generate each term by doubling the previous term and adding 1. Sequence: 3, 7, 15, 31, 63, ...There are two main ways to compute the $n$-th term computationally:1. **Iterative (Bottom-up):** Start at term 1, apply the rule $n-1$ times using a loop.2. **Recursive (Top-down):** Define a function that calls *itself*. The function asks for term $n-1$, which asks for term $n-2$, until hitting the **Base Case** (term 1).

### Iterative vs Recursive TracingSuppose we want the 3rd term of the sequence.**Iterative:**- `term = 3` (1st)- `term = 2*3 + 1 = 7` (2nd)- `term = 2*7 + 1 = 15` (3rd) -> Return 15.**Recursive:**- `f(3)` returns `2 * f(2) + 1`  - `f(2)` returns `2 * f(1) + 1`    - `f(1)` hits the base case: returns `3`  - `f(2)` gets 3, computes `2*3+1 = 7`, returns `7`- `f(3)` gets 7, computes `2*7+1 = 15`, returns `15`.

### 🔍 ReflectionWhy is the recursive Fibonacci so absurdly slow? <details><summary>💡 Answer</summary>Because `fib(35)` calls `fib(34)` and `fib(33)`. But `fib(34)` *also* calls `fib(33)` and `fib(32)`. The number of function calls doubles at each step, resulting in $O(2^n)$ time complexity!
