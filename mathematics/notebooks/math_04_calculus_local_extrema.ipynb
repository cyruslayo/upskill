---## Phase −1 — Theoretical Foundation### Finding Local Extrema> 📖 *From Calculus §1.8:* Derivatives can be used to find a function’s **local extreme values**, its peaks and valleys. At its peaks and valleys, a function’s derivative is either $0$ (a smooth, rounded peak/valley) or undefined (a sharp, pointy peak/valley).### Critical PointsThe points at which a function’s derivative is $0$ or undefined, and the function itself exists, are called **critical points**. We find them by:1. Taking the derivative $f'(x)$2. Noting any singularities (where $f'(x)$ is undefined)3. Setting the derivative to $0$ and solving for $x$**Example:** $f(x) = x^3 - 3x^2$1. $f'(x) = 3x^2 - 6x$2. No singularities.3. $3x^2 - 6x = 0 \Rightarrow 3x(x - 2) = 0 \Rightarrow x = 0, x = 2$So the critical points are $x=0$ and $x=2$.

```python
# Finding critical points symbolicallyimport sympy as spx = sp.Symbol('x')f = x**3 - 3*x**2f_prime = sp.diff(f, x)critical_points = sp.solve(sp.Eq(f_prime, 0), x)print(f'Critical points of {f}: {critical_points}')
```

### Classifying Critical Points (First Derivative Test)Not all critical points are peaks or valleys; some are **saddle points** (flat terrain on the side of a mountain).To classify a critical point $c$, inspect the sign of $f'(x)$ on either side:- If $f'$ changes from **+ to -**, it's a **local maximum** (ascending, then descending)- If $f'$ changes from **- to +**, it's a **local minimum** (descending, then ascending)- If $f'$ does not change sign, it's a **saddle point** (neither)
