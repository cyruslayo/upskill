---## Phase −1 — Theoretical Foundation### The Idea Behind Euler's Method> 📖 *From Algorithms Ch. 18:* Arrays can be used to implement mathematical procedures like Euler estimation to solve differential equations.If we know the derivative $f'(x, y)$ (the slope), and we know a starting point $(x_0, y_0)$, we can approximate the next point by taking a tiny step $\Delta x$ along the tangent line:$$y_{n+1} = y_n + f'(x_n, y_n) \cdot \Delta x$$$$x_{n+1} = x_n + \Delta x$$By chaining these tiny steps together, we can trace out the entire curve $y(x)$ without ever actually integrating the function analytically!### Multivariable SystemsIf we have a *system* of differential equations, e.g., $$a'(t) = a(t) + 1$$$$b'(t) = a(t) + b(t)$$We simply update *all* variables simultaneously at each step:$$a_{n+1} = a_n + a'(t, a_n, b_n) \cdot \Delta t$$$$b_{n+1} = b_n + b'(t, a_n, b_n) \cdot \Delta t$$$$t_{n+1} = t_n + \Delta t$$

This uses the slope at the *beginning* of the interval.Explain why this "forward" Euler method systematically overestimates or underestimates curves that are concave up or concave down. How could we improve it?<details><summary>💡 Sketch after a retrieval attempt</summary>If a curve is bending upwards (concave up), its slope is constantly increasing. But forward Euler locks in the slope at the start of the interval, extending a straight line. Because the true curve bends up *away* from this line, Euler systematically underestimates the curve.A better approach is the Runge-Kutta (RK4) method, which calculates the slope at the beginning, midpoint, and end of the interval, and takes a weighted average to predict the curve's bend!</details>---📚 **Next:** Algorithms 19 (SIR Epidemic Model)
