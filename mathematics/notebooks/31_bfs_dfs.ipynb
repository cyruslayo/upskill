---## Phase −1 — Theoretical Foundation### Graph Representations> 📖 *From Algorithms Ch. 31:* A graph is a set of vertices (nodes) and edges (connections). Edges can be directed or undirected.To represent a graph in code, we generally use an **Adjacency List** (a dictionary mapping each node to a list of its neighbors) or an **Adjacency Matrix** (a 2D grid where `matrix[i][j]` is 1 if an edge exists, 0 otherwise).### Graph Traversal: BFS vs DFSTo explore a graph, we need a systematic way to visit all reachable nodes without getting stuck in infinite loops.- **Breadth-First Search (BFS)**: Explores all neighbors of a node before moving deeper. It ripples outwards like a wave. It is guaranteed to find the *shortest path* in an unweighted graph. It uses a **Queue** (First-In, First-Out).- **Depth-First Search (DFS)**: Plunges as deep as possible down one path before backtracking. It is great for maze-solving and detecting cycles. It uses a **Stack** (Last-In, First-Out), which can be managed explicitly or via the recursive call stack.

### Comprehension Check ✅If you want to find the fastest route out of a maze, should you use BFS or DFS? What if you want to find ALL possible routes out of the maze?<details><summary>💡 Explanation after a retrieval attempt</summary>Use **BFS** to find the absolute fastest route, because it naturally explores shorter paths before longer ones. Use **DFS** to exhaustively find all possible routes, because it easily traces single paths from start to finish and then backtracks.</details>
