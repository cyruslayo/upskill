---## Phase −1 — Theoretical Foundation### Linear-Encoding Cryptography> 📖 *From Algorithms Ch. 7:* An encoding function maps a string to a sequence of numbers. Using a linear encoding function $f(x) = ax + b$, we can encode messages.Suppose $a=2$ and $b=3$. The letter 'a' is $x=1$. $f(1) = 2(1) + 3 = 5$.The space ' ' is $x=0$. $f(0) = 2(0) + 3 = 3$.### DecodingTo decode, we invert the function: $x = \frac{f(x) - b}{a}$.If we receive $5$, $x = \frac{5 - 3}{2} = 1$ ('a').### Brute-Force SearchIf we don't know $a$ and $b$, but we know they are small integers, we can try *every* possible combination of $(a, b)$. This is **brute-force search**.We check each decoded sequence. If all characters are valid integers between 0 and 26, it *might* be a valid message. If it contains real words, we've broken the code!

### Comprehension Check ✅1. If the encoded message is `[7, 3, 11]` and $a=2, b=1$, what is the decoded sequence of numbers?2. What is the fundamental weakness of brute-force search if $a$ and $b$ can be any 64-bit integer?<details><summary>💡 Explanation after a retrieval attempt</summary>1. $x = \frac{y-1}{2}$. $7 \rightarrow 3$, $3 \rightarrow 1$, $11 \rightarrow 5$.
