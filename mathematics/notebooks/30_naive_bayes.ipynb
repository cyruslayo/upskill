---## Phase −1 — Theoretical Foundation### Bayesian Probability> 📖 *From Algorithms Ch. 30:* Naive Bayes is grounded in Bayesian probability. It chooses the class with the highest conditional probability given the data point, assuming that all the variables in the dataset are independent from each other.From Bayes' Theorem: $$P(\text{class} \mid \text{point}) = \frac{P(\text{point} \mid \text{class}) P(\text{class})}{P(\text{point})}$$Since $P(\text{point})$ is the same for all classes, we just want to maximize the numerator: $P(\text{point} \mid \text{class}) P(\text{class})$.### The "Naive" AssumptionA point has multiple features $(x_1, x_2, \dots)$. Calculating the exact joint probability $P(x_1, x_2 \mid \text{class})$ is nearly impossible. But if we Naively assume the features are strictly independent, we can multiply them!$$P(x_1, x_2 \mid \text{class}) \approx P(x_1 \mid \text{class}) \times P(x_2 \mid \text{class})$$Thus, the formula becomes:$\text{Score} = P(\text{class}) \times P(x_1 \mid \text{class}) \times P(x_2 \mid \text{class})$The algorithm computes this score for every possible class, and predicts the class with the highest score!

### Comprehension Check ✅Why is the independence assumption called "Naive"?<details><summary>💡 Explanation after a retrieval attempt</summary>Because features in the real world are almost never perfectly independent! For example, if predicting whether an animal is a Dog, the feature "Has Fur" and "Barks" are heavily correlated. Treating them as independent statistical events is a mathematical fiction, but it turns out to work remarkably well in practice anyway.</details>
