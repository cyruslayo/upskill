---## Phase −1 — Theoretical Foundation### Multivariable Gradient Descent> 📖 *From Algorithms Ch. 11:* Multivariable gradient descent is similar to single-variable, except we replace the scalar derivative $f'(x)$ with the **gradient vector** $\nabla f(\vec{x})$.$$\vec{x}_{n+1} = \vec{x}_n - \alpha \nabla f(\vec{x}_n)$$If $f$ is a function of $(x, y)$, the gradient is a vector containing the partial derivatives:$$\nabla f(x, y) = \left\langle \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right\rangle$$For example, if $f(x, y) = x \sin(y) + x^2$:- $\frac{\partial f}{\partial x} = \sin(y) + 2x$- $\frac{\partial f}{\partial y} = x \cos(y)$Our update rule becomes two simultaneous equations:$$x_{n+1} = x_n - \alpha (\sin(y_n) + 2x_n)$$$$y_{n+1} = y_n - \alpha (x_n \cos(y_n))$$
