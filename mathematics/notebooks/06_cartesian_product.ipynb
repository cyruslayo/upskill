---## Phase −1 — Theoretical Foundation### What is a Cartesian Product?> 📖 *From Algorithms Ch. 6:* The Cartesian product constructs all the points whose elements occupy the given ranges. For example, if we have ranges `[1, 2]` and `['A', 'B']`, the Cartesian product is:`[[1, 'A'], [1, 'B'], [2, 'A'], [2, 'B']]`.### The Algorithm1. Create an array initialized with an empty point: `points = [[]]`.2. For each range in your input:   - Create a `new_points` array.   - For every point currently in `points`, and every item in the current range, append the item to the point and save it to `new_points`.   - Update `points = new_points`.### The Array Copying TrapIn Python, if `a = [1]`, doing `b = a` does NOT copy the array. Both `a` and `b` point to the exact same list in memory. Doing `b.append(2)` will modify `a` as well!To make a new copy, you must use `b = a[:]` or `b = a.copy()` or `b = list(a)`.
