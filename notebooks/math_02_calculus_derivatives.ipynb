## Concept Introduction
From *Calculus* Sections 1.3 & 1.4.
The derivative represents the instantaneous slope of a function. 
The **Difference Quotient** approximates it: `f'(x) ≈ (f(x + dx) - f(x)) / dx`.
The **Power Rule** provides an exact symbolic formula: if `f(x) = cx^n`, then `f'(x) = c * n * x^(n-1)`.
